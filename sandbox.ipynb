```python
import pandas as pd
```

```python
inquiry_df = pd.read_csv('./data/DetailInquiry_20200130160108.csv')
```

```python
inquiry_df.head()
```

```python
trade_df = pd.read_csv('./data/SaveFile_000001_000050.csv')
```

```python
trade_df.head()
```

```python
balance_df = pd.DataFrame(columns=['date', 'reason', 'brand', 'amount'])
```

```python
for inquiry in inquiry_df.itertuples():
    if inquiry.摘要.startswith('株式配当金'):
        balance_df = balance_df.append({
            'date': inquiry.入出金日,
            'reason': 'dividend',
            'brand': inquiry.摘要.split(' ')[1],
            'amount': int(inquiry.入金額)
        }, ignore_index=True)
    elif inquiry.摘要 == '譲渡益税源泉徴収金':
        balance_df = balance_df.append({
            'date': inquiry.入出金日,
            'reason': 'duty',
            'brand': '',
            'amount': -int(inquiry.出金額)
        }, ignore_index=True)
    else:
        print('Unknown inquiry type!!: ' + inquiry.摘)
```

```python
trade_df
```

```python
stock = {}
for trade in trade_df.itertuples():
    if trade.取引 == '株式現物買':
        if trade.銘柄コード in stock:
            stock[trade.銘柄コード] = {
                "number": stock[trade.銘柄コード]["number"] + trade.約定数量,
                "price": stock[trade.銘柄コード]["price"] + trade.約定数量 * trade.約定単価
            }
        else:
            stock[trade.銘柄コード] = {
                "number": trade.約定数量,
                "price": trade.約定数量 * trade.約定単価
            }
    elif trade.取引 == '株式現物売':
        if trade.銘柄コード in stock:
            stock[trade.銘柄コード] = {
                "number": stock[trade.銘柄コード]["number"] - trade.約定数量,
                "price": stock[trade.銘柄コード]["price"] - trade.約定数量 * trade.約定単価
            }
            if stock[trade.銘柄コード]["price"] < 0:
                balance_df = balance_df.append({
                    'date': trade.約定日,
                    'reason': 'trade',
                    'brand': trade.銘柄,
                    'amount': -stock[trade.銘柄コード]["price"]
                }, ignore_index=True)
    elif trade.取引 == '投信金額買付':
        print('TODO ' + trade.取引)
    else:
        print('Unknown trade type!!: ' + trade.取引)
```

```python
balance_df['date'] = pd.to_datetime(balance_df['date'])
balance_df.set_index('date', inplace=True)
```

```python
balance_df = balance_df.sort_values('date')
```
